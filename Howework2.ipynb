# Условие 3: 3 задача

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

```python
plt.figure(figsize=(8,6))
plt.hist(df['bedrooms'],bins=32)
plt.title('Количество спален в квартире')
plt.xlabel('Колл-во спален')
plt.ylabel('Кол-во')
plt.yticks(range(0,10000,1000))
plt.xticks(range(0,31,5),rotation=1)
plt.grid();
```

По данному графику можно сказать, что рынок недвижимости перенасыщен квартирами с 3 спальнями (около 9500 продается квартир)

```python
sns.boxplot(x=df['bedrooms'],y=df['grade']);
```

График показывает, что оценка квартир с 7 спальнями в среднем достаточно высокая. У квартир на 3 спальни, которыми рынок недвижимости достаточно насыщеный, оценка достаточно ниже, и в среднем ранжируется от 7 до 9.

```python
sns.boxplot(x=df['bedrooms'],y=df['price']);
```

Самые дорогие на рынке - это квартиры на 8 спален, при этом встречаются отдельные квартиры 5 и 6 спален, где цена завышена. Квартиры на 3 спальни цена ниже 1.000.000.

```python
sns.boxplot(x=df['waterfront'],y=df['price']);
```

Цена на квартиры возрастает, если находится рядом с берегом.

```python
data_corr = df[['price','bedrooms','waterfront','grade']].corr()
data_corr
```

```python
sns.heatmap(data_corr, annot=True, cmap='coolwarm', linewidths=1);
```

В среднем цена недвижимости на рынке зависит от оценки жилья и количества спален в квартире. 

Вывод: На рынке недвижимости сейчас преобладают квартиры с количеством спален на 3. Оценка таких квартир средняя. Цена в сравнение с квартирами, начиная с 4х спален, варьируется от 300.000 до 500.000. Цена на другие же квартиры начинает увеличиваться от увелечения числа спален и высокой оценки жилья. За некоторые квартиры цена завышена из-за близкого расположения к берегу.
